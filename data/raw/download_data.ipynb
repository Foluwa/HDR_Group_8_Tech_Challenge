```python
def download_raw_data(dataset: str, file_id: str):
    file_name = f"{dataset}.csv"

    if file_name not in os.listdir():
        gdown.download(id=file_id, output=file_name, quiet=False)
```

```python
with ThreadPoolExecutor() as executor:
    results = list(executor.map(lambda item: download_raw_data(*item), datasets.items()))
```
